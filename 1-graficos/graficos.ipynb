Gráficos


```python
# Importar as bibliotecas necessárias para geração dos gráficos e dos números
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Criar 100 dados igualmente intervalados entre 0 e 2
x = np.linspace(0, 2, 100)

# Função Linear
y1 = x*2

# Função Quadrática
y2 = x**2
```

```python
# Plotar dados
plt.plot(x, y1, "*r")
plt.plot(x, y2, "g--")

# Legenda nos eixos
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

# Título do gráfico
plt.title("Título do gráfico")
```

```python
# Função Linear 
y1 = x*2

# Função quadrática
y2 = x**2

# Função cúbica
y3 = x**3

# Função senoidal
y4 = np.sin(x)
```

```python
# Plots lado a lado 
# Gerar janela com 2 linhas e 2 colunas
fig, janela = plt.subplots(2, 2)

# Adicionar a cada janela um gráfico 
janela[0, 0].plot(x, y1, 'r')
janela[0, 1].plot(x, y2, 'g')
janela[1, 0].plot(x, y3, 'b')
janela[1, 1].plot(x, y4, 'y')

# Adicionar legenda 
fig.legend(['Linear', 'Quadrática', 'Cúbica', 'Senoidal'])
```

```python
from matplotlib import colors
# Desenhar gráfico
plt.plot([1, 2, 3, 4], [10, 15, 22, 31],
         linestyle='--', color='red', marker='s', linewidth='3.0')

# Ajustar eixos do gráfico
plt.axis([0, 10, 0, 40]) # [x1, x2, y1, y2] Definindo limites di plano cartesiano
plt.legend('Gráfico')
plt.title('Título do gráfico')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```

## Gráfico de barra

Podemos obter um gráfico de barras com a função
bar()

```python
# Valores para o Bar Chart
eixo_y = [10, 30, 15]
eixo_x = range(len(eixo_y))

largura = 0.5
cor_barra = '#00ff00'

plt.bar(eixo_x, eixo_y, width=largura, color=cor_barra)
plt.show()
```

## Histogramas

```python
# Gerar números aleatórios com didtribuição normal
num_norm = np.random.randn(1000, 1) # 1000 linhas e 1 coluna
```

```python
# bins - representa a quantidade de caixas com intervalos de observação de frequência
plt.hist(num_norm, bins=10)
plt.xlabel('Valores')
plt.ylabel('Frequência Observada')
plt.show()
```
